```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau = 5 # Period
s = 0 # Shift
r = 1 # Rateo of Signal blocked
```

```python
def phi_f(x):
    return ((x-s) % tau ) / tau
```

```python
def k_f(x):
    if x <= r/2:
        return 2*x/r
    elif r/2 < x <= r:
        return 2 - 2*x/r
    else:
        return 0
```

```python
t = np.arange(0, 10, 0.01)
phi = phi_f(t)
k = np.array([k_f(x) for x in phi])
```

```python
plt.plot(t, phi, c='r', alpha=0.1, label="phi(t)")
plt.plot(t, k, c='b', alpha=0.5, label="k(t)")
# plt.show()
plt.plot()
plt.legend()
plt.title("tau={}, s={}, r={}".format(tau, s, r))
plt.savefig("plot.png")
```
